# Model Interpretation Results Summary

This notebook provides a summary of the model interpretation results generated by the `interpret_model.py` script. It includes global feature importance and detailed local explanations for specific examples.

```python
import pandas as pd
from IPython.display import display, Image
from pathlib import Path

# Define the base path for results relative to this notebook
results_base_path = Path('../results/interpretation')

print("[OK] Libraries imported and paths set.")
```

## ðŸ“Š Global Feature Importance (SHAP Summary)

The SHAP summary plot provides an overview of the global feature importance, showing which features have the most impact on model predictions and the direction of their effect.

```python
shap_summary_plot_path = results_base_path / 'shap_summary_plot.png'
if shap_summary_plot_path.exists():
    display(Image(filename=shap_summary_plot_path))
    print(f"[OK] Displayed: {shap_summary_plot_path}")
else:
    print(f"[ERROR] SHAP summary plot not found at {shap_summary_plot_path}")
```

## ðŸ“‹ Local Explanations (5 Examples)

The model interpretation script identified 5 key examples (High Risk, Safe, and Borderline) and generated detailed local explanations (waterfall plots) along with a summary table of their probabilities and top contributing features.

```python
local_analysis_csv_path = results_base_path / 'local_examples_analysis.csv'
if local_analysis_csv_path.exists():
    local_df = pd.read_csv(local_analysis_csv_path)
    print("### Summary of Local Examples:")
    display(local_df)
    print(f"[OK] Loaded and displayed: {local_analysis_csv_path}")
else:
    print(f"[ERROR] Local examples analysis CSV not found at {local_analysis_csv_path}")
```
